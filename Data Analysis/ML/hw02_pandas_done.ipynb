#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
uploaded = files.upload()

zip_filename = next(iter(uploaded.keys()))
with zipfile.ZipFile(io.BytesIO(uploaded[zip_filename]), 'r') as zip_ref:
    zip_ref.extractall('Data')

students_info = pd.DataFrame()
students_marks = pd.DataFrame()
data_folder = 'Data/Data'

for file_name in os.listdir(data_folder):
  if file_name.startswith("Students_info"):
    df = pd.read_csv(os.path.join(data_folder, file_name))
    students_info = pd.concat([students_info,df],ignore_index=True)
  else:
    df = pd.read_csv(os.path.join(data_folder, file_name))
    students_marks = pd.concat([students_marks,df],ignore_index=True)

df = pd.merge(students_info, students_marks, on='index').sort_values(by='index')
df
```

```python
start_time = time.time()

min_scores = {}

educations = df['education'].unique()

for education in educations:
    filtered_df = df[df['education'] == education]
    min_score = filtered_df['writing score'].min()
    min_scores[education] = min_score

df2 = pd.DataFrame({'education': list(min_scores.keys()), 'min_writing_score': list(min_scores.values())})

end_time = time.time()
execution_time = end_time - start_time
print(df2)
print(f'Время выполнения программы: {execution_time}')
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
# your code here
df['Math_PassStatus'] = df['math score'].apply(lambda x: 'P' if x >= passmark else 'F')

math_pass_counts = df['Math_PassStatus'].value_counts()

print(f"Количество студентов, сдавших и не сдавших экзамен по математике:")
print(math_pass_counts)

df['Reading_PassStatus'] = df['reading score'].apply(lambda x: 'P' if x >= passmark else 'F')

df['Writing_PassStatus'] = df['writing score'].apply(lambda x: 'P' if x >= passmark else 'F')

reading_pass_counts = df['Reading_PassStatus'].value_counts()
writing_pass_counts = df['Writing_PassStatus'].value_counts()

print("\nКоличество студентов, сдавших и не сдавших экзамен по чтению:")
print(reading_pass_counts)
print("\nКоличество студентов, сдавших и не сдавших экзамен по письму:")
print(writing_pass_counts)
df.head(5)
```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

Посчитайте количество студентов, которые сдали все экзамены.

```python
# your code here
df['OverAll_PassStatus'] = df.apply(lambda row: 'P' if row['Math_PassStatus'] == 'P' and
                                    row['Reading_PassStatus'] == 'P' and
                                    row['Writing_PassStatus'] == 'P' else 'F', axis=1)

overall_pass_counts = df['OverAll_PassStatus'].value_counts()

print("Количество студентов, успешно сдавших все экзамены:")
print(overall_pass_counts)
df.head(5)
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark > 90:
        return 'A'
    elif 80 <= average_mark <= 90:
        return 'B'
    elif 70 <= average_mark < 80:
        return 'C'
    elif 60 <= average_mark < 70:
        return 'D'
    elif 50 <= average_mark < 60:
        return 'E'
    else:
        return 'F'

df['Grade'] = df[['math score', 'reading score', 'writing score']].mean(axis=1).apply(GetGrade)

grade_counts = df['Grade'].value_counts()

print("Количество студентов, получивших каждую оценку:")
print(grade_counts)
df.head(5)
```
